```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline

earthquakes_df = pd.read_csv('earthquakes_data_Feb_2016.csv')
earthquakes_df.head()

import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize=(4, 3))
fig = plt.scatter(x=earthquakes_df['properties.cdi'], 
            y=earthquakes_df['properties.mag'], 
           )
plt.xlabel('CDI')
plt.ylabel('Magnitude')
```

```python
import seaborn as sns
sns.set_context("paper", font_scale=1.1)
plt.figure(figsize=(4, 3))
# show adding the whitegrid, removing it. Show adding ticks
sns.set_style("darkgrid")
ax = sns.regplot(x='properties.cdi', 
            y='properties.mag', 
            fit_reg=False,
           data=earthquakes_df)
ax.set(xlabel='CDI', ylabel='Magnitude')
```

```python
plt.figure(figsize=(4, 3))
sns.set_style("whitegrid")
sns.regplot(x='properties.cdi', 
            y='properties.mag', 
            fit_reg=False,
           data=earthquakes_df)
```

```python
plt.figure(figsize=(4, 3))
sns.set_style("white")
sns.regplot(x='properties.cdi', 
            y='properties.mag', 
            fit_reg=False,
           data=earthquakes_df)
sns.despine()
```

```python
plt.figure(figsize=(4, 3))
sns.set_style("dark")
sns.regplot(x='properties.cdi', 
            y='properties.mag', 
            fit_reg=False,
           data=earthquakes_df)
```

```python
plt.figure(figsize=(4, 3))
sns.set_style("ticks")
sns.regplot(x='properties.cdi', 
            y='properties.mag', 
            fit_reg=False,
           data=earthquakes_df)
sns.despine()
```

```python
plt.figure(figsize=(4, 3))

plot = sns.regplot(x='properties.cdi', 
            y='properties.mag', 
            fit_reg=False,
           data=earthquakes_df)
plot.set(xlabel='CDI', ylabel='Magnitude')
plot.get_xaxis().set_ticks([4, 8])
plot.get_yaxis().set_ticks([4,8])
sns.despine()
sns.plt.xlim(0,)
sns.plt.ylim(0,)
```
